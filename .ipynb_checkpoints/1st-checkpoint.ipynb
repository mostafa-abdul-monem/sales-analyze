```python
L1 = [1,2,4,2,6,9,88,11,22,4,33,6,8,9]
print(f"Maximum value: {max(L1)}")
print(f"Minimum value: {min(L1)}")
print(f"The sum is {sum(L1)}")
mean = sum(L1) / len(L1)
print(f"The mean is {mean}")
```

```python
sortedL1 = sorted(L1)
sortedL1
```

```python
if (len(sortedL1) % 2) == 1:
    print('the median is ', sortedL1[len(sortedL1) // 2])
else:
    print('the median is ', (sortedL1[len(sortedL1)// 2] + sortedL1[len(sortedL1)// 2 -1]) / 2)
```
